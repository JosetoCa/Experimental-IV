## TEORÍA.
Siguiendo lo expuesto en [], El efecto fotoeléctrico se refiere a la emisión de electrones de una superficie cuando se hace incidir luz sobre esta.  Para escapar de una superficie, el electrón debe absorber energía suficiente de la radiación incidente para superar la atracción de los iones positivos del material de la superficie. La cantidad mínima de energía que debe ganar un electrón para escapar de determinada superficie se llama función trabajo para esa superficie, y se representa con $\phi$. Siguiendo el postulado de Planck sobre la luz viajando en cuantos, fotones, la cantidad de energía que un fotón le puede entregar a un electrón es:
$$
E = h\nu
$$
Donde $h$ es la constante de Planck y $\nu$ es la frecuencia del fotón.

Un fotón que llega a la superficie es absorbido por un electrón. Esta transferencia de energía es un proceso de todo o nada, a diferencia de la transferencia continua de energía de la teoría clásica; el electrón obtiene toda la energía del fotón o no obtiene nada. Si esta energía es mayor que la función trabajo $\phi$, el electrón puede escapar de la superficie. Cuando es mayor la intensidad a determinada frecuencia equivale a una cantidad proporcionalmente mayor de fotones que se absorben por segundo y, en consecuencia, resulta una cantidad proporcionalmente mayor de electrones emitidos por segundo.
Con base en estos principios la energía cinética, $E_c$ con la que sale disparado un electrón del material es:
$$
E_c  = h\nu - \phi
$$
Mientras la energía es negativa el electrón permanece ligado al material.

## Montaje
Para el *primer* acercamiento experimental a la constante de Planck se usó una celda fotoeléctrica: un tubo de vacío, de alrededor de ($0.01$ Pa) con dos electrodos conductores, cátodo y ánodo, que se conectan a una fuente de diferencia de potencial, formando un campo eléctrico cuya dirección es de cátodo a ánodo.  La luz que llega a la superficie del cátodo crea una corriente en el circuito externo, si los electrones tienen suficiente energía potencial; es decir, el campo eléctrico hace los electrones  emitidos del material sufran un proceso de frenado. Una vez la diferencia de potencial es mínima, frenan los electrones emitidos y la energía dada por el campo eléctrico al electrón, $E_c$, debe ser igual a su energía cinética. Es decir:
$$
E_E = qV = E_c = h\nu - \phi 
$$
Donde $q$ es la carga del electrón y $V$ es el voltaje de frenado

Mientras que la energía del potencial está determinada por la menor frecuencia que puede tener un fotón, $\nu_0$, para que ocurra el efecto fotoeléctrico. Por tanto, en el montaje propuesto:
$$
qV = h(\nu-\nu_0)
$$
Se hizo una medición del voltaje 

Utilizando la ecuación de Einstein:
$$
qV = h(\nu -\nu_0) 
$$


```python
import numpy as np
from scipy.stats import linregress
from scipy.constants import e, c, h
import matplotlib.pyplot as plt


emision = np.array([454, 520, 640])*1e-9 # En m
voltajes = np.array ([0.976, 0.758, 0.3188])


slope, intercept, r_value, p_value, std_err = linregress(c/emision, voltajes)

plt.scatter(c/emision, voltajes, s=2, label = 'Datos Experimentales')
plt.plot(c/emision, slope*c/emision + intercept, label = 'ajuste')
plt.legend()
plt.grid()
plt.show()


print(slope)
print(r_value**2)
print(abs(slope-(h/e))/(h/e)*100)
```

Vamos a ajustar datos de voltaje y longitud de onda a la siguiente ecuación:
$$
qV = h(\nu-\nu_0)
$$
$$
qV = h\nu-h\nu_0
$$

$$
\frac{qV}{\frac{c}{\lambda}-\frac{c}{\lambda_0}} = h
$$
$$
\frac{qV\lambda \lambda_0}{c\left({\lambda_0}-{\lambda}\right)} = h
$$

```python
import numpy as np
from scipy.stats import linregress
from scipy.constants import e, c, h
import matplotlib.pyplot as plt


emision = np.array([451.87, 526.54, 642.91])*1e-9 # En m
voltajes = np.array ([2.175, 1.8, 1.425])


slope, intercept, r_value, p_value, std_err = linregress(c/emision, voltajes)

plt.scatter(c/emision, voltajes, s=2, label = 'Datos Experimentales')
plt.plot(c/emision, slope*c/emision + intercept, label = 'ajuste')
plt.legend()
plt.grid()
plt.show()


print(slope)
print(r_value**2)
print(h/e)
print(abs(slope-(h/e))/(h/e)*100)

```
