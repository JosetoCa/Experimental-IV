```python
import numpy as np
from scipy.stats import linregress
from scipy.constants import e, c, h
import matplotlib.pyplot as plt


### DATOS EXPERIMENTALES
#LONGITUD DE ONDA DE LOS LÁSERES EN METROS
emision = np.array([454, 520, 640])*1e-9 # En m
#VOLTAJES DE FRENADO EN VOLTIOS
voltajes = np.array ([0.976, 0.758, 0.3188])

#USO DE LA FUNCIÓN LINREGRESS DE SCIPY
slope, intercept, r_value, p_value, std_err = linregress(c/emision, voltajes) #EN EL EJE X SE PONEN LOS DATOS DE FRECUENCA = C/LAMBDA

#GRÁFICA DE LOS DATOS
plt.scatter(c/emision, voltajes, s=10, label = 'Datos Experimentales', color = 'red')
plt.plot(c/emision, slope*c/emision + intercept, label = 'Regresión', color = 'green')
plt.title('Gráfica de datos experimentales primer montaje.')
plt.xlabel('frecuencia [Hz]')
plt.ylabel('Voltaje [V]')
plt.legend()
plt.grid()
plt.show()

#SE IMPRIMEN LOS RESULTADOS DE LA FUNCIÓN LINREGRESS 
print(slope)
#ERROR RELATIVO DE LA PENDIENTE CON LA CONSTANTE DE PLANCK EN eV.
print(abs(slope-(h/e))/(h/e)*100)
print(intercept)
print(std_err)
```
