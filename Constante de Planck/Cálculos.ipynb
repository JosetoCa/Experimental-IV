Vamos a ajustar datos de voltaje y longitud de onda a la siguiente ecuaci√≥n:
$$
qV = h(\nu-\nu_0)
$$
$$
qV = h\nu-h\nu_0
$$

$$
\frac{qV}{\frac{c}{\lambda}-\frac{c}{\lambda_0}} = h
$$
$$
\frac{qV\lambda \lambda_0}{c\left({\lambda_0}-{\lambda}\right)} = h
$$

```python
import numpy as np
from scipy.stats import linregress
from scipy.constants import e, c
import matplotlib.pyplot as plt


emision = np.array([451.87, 526.54, 642.91])*1e-9 # En m
voltajes = np.array ([2.175, 1.8, 1.425])/c


slope, intercept, r_value, p_value, std_err = linregress(1/emision, voltajes)

plt.scatter(1/emision, voltajes, s=2, label = 'Datos Experimentales')
plt.plot(1/emision, slope*1/emision + intercept, label = 'ajuste')
plt.legend()
plt.grid()
plt.show()


print(slope*e)
print(r_value**2)

```
