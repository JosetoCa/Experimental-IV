El coeficiente de dilatación térmica determinado en esta práctica es una medida indirecta de la longitud de las varillas a una temperatura inicial, y el correspondiente cambio de longitud debido a un cambio de temperatura. Asumiendo pequeñas incertidumbres asociadas a la precisión de los instrumentos, la incertidumbre de el coeficiente de dilatación viene siendo:

$$
\delta \alpha = \sqrt{\left( \frac{\partial \alpha}{\partial L_0} * \delta L_0 \right)^2 + \left( \frac{\partial \alpha}{\partial \Delta L} * \delta \Delta L \right)^2 + \left( \frac{\partial \alpha}{\partial \Delta T} * \delta \Delta T  \right)^2}
$$


$$
\delta \alpha = \sqrt{\left( \frac{\Delta L}{{L_0}^2 \Delta T} * \delta L_0 \right)^2 + \left( \frac{1}{L_0 \Delta T} * \delta \Delta L \right)^2 + \left( \frac{\Delta L}{L_0 \Delta T ^2 } * \delta \Delta T  \right)^2}
$$

$$
\delta \alpha = \sqrt{\left( \frac{\delta L_0}{L_0} * \alpha \right)^2 + \left( \frac{\delta \Delta L}{\Delta L} * \alpha \right)^2 + \left( \frac{\delta \Delta T}{\Delta T } * \alpha  \right)^2}
$$

$$
\delta \alpha = \alpha \sqrt{\left( \frac{\delta L_0}{L_0}  \right)^2 + \left( \frac{\delta \Delta L}{\Delta L}  \right)^2 + \left( \frac{\delta \Delta T}{\Delta T }  \right)^2}
$$

```python
import numpy as np
def coe_dil(dl,l,dt):
    return dl/(l*dt)
def incer(alfa,dl,l,dt,il,idl,idt):
    return alfa*np.sqrt((il/l)**2 + (idl/dl)**2 + (idt/dt)**2)
idl = .01
il = 1
idt = .1
```

```python
L_f = 701
dl_f = .59
T_0 = 27
T_f1 = 86.45
dt = T_f1-T_0
print("Bajo la aproximación de promediar las temperaturas finales de los sensores el coeficiente de dilatación térmica del hierro fue: "+f"{coe_dil(dl_f, L_f, dt):.2e}" + r' $\pm$ '  + f"{incer(coe_dil(dl_f, L_f, dt), dl_f, L_f, dt, il, idl, idt ):.0e}")
T_f1 = 88.2
dt = T_f1-T_0
print("Bajo la aproximación de asumir la mayor temperatura medida como la final el coeficiente de dilatación térmica del hierro fue: "+f"{coe_dil(dl_f, L_f, dt):.2e}" + r' $\pm$ '  + f"{incer(coe_dil(dl_f, L_f, dt), dl_f, L_f, dt, il, idl, idt ):.0e}")

```
