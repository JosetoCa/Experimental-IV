$$
k = \frac{q}{nbT} \pm \sqrt{\left(\frac{\partial k}{\partial T}\delta T\right)^2+\left(\frac{\partial k}{\partial b}\delta b\right)^2}
$$

$$
k = \frac{q}{nbT} \pm \sqrt{\left( \frac{q}{nbT^2} \delta T\right)^2+\left(\frac{q}{nTb^2}\delta b\right)^2}
$$
$$
\delta k = \frac{q}{bnT}*\sqrt{(b\delta T)^2 + (T\delta b)^2}
$$

$$
\delta k = \frac{k}{bT}*\sqrt{(b\delta T)^2 + (T\delta b)^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.constants import e


voltajes = [voltaje_1, voltaje_2, voltaje_3, voltaje_4]
corrientes = [corriente_1, corriente_2, corriente_3, corriente_4]
temperaturas = [temperatura1, temperatura2, temperatura3, temperatura4]

def exponencial(x, a, b):
    return a * np.exp(b * x)
b = []
a = []
b_error = []
a_error = []
T_error = []

#si vamos a hacer lo de chi
corrientes_error = [np.ones_like(corrientes[0]*0.001),np.ones_like(corrientes[1]*0.01),np.ones_like(corrientes[2]*0.001),np.ones_like(corrientes[3]*0.01)]
for i in range(len(voltajes)):
    corrientes_error.append(np.ones_like(corrientes[i]) * 0.001) # Error de 1 mV en los voltajes

chis = []



for i in range(len(voltajes)):
    popt, pcov = curve_fit(exponencial, voltajes[i],corrientes[i], p0=[0, 0])
    b.append(popt[1])
    a.append(popt[0])
    b_error.append(np.sqrt(np.diag(pcov))[1])
    a_error.append(np.sqrt(np.diag(pcov))[0])
    T_error.append(0.1)
    chis.append(np.sum(((corrientes[i] - exponencial(voltajes[i], popt[0], popt[1])) / corrientes_error[i])**2))



k = []
k_error = []
ideality = [1.8, 1.8, 1.7, 1.7]  

for i in range(len(temperaturas)):
    k.append((1/ideality[i])* (e) / (temperaturas[i] * b[i]))
    k_error.append(k[i]/(b[i]*temperaturas[i]) * np.sqrt((b_error[i]*temperaturas[i])**2 + (T_error[i]*b[i])**2))


k_error = np.array(k_error)

k_boltzmann = 0
k_boltzmann_error = 0
total_datos = 0
for i in range(len(k)):
    total_datos += len(voltajes[i])


for i in range(len(k)):
    k_boltzmann_error += (len(voltajes[i]) * k_error[i] / total_datos)**2

k_boltzmann_error = np.sqrt(k_boltzmann_error)



for i in range(len(k)):
    k_boltzmann += k[i] * len(voltajes[i]) / total_datos

print("k_boltzmann_error:", k_boltzmann_error)
print("k_boltzmann:", k_boltzmann)
print(chis)
```

Pensando en linealizar, nuestro modelo es tal que:
$$
I = ae^{Vb}
$$
con $b = \frac{q}{T K_b N}$. Sacando logaritmo a ambos lados:

$$
\ln{I} = Vb + \ln{a}
$$

La pendiente de un ajuste lineal será el valor relacionado con la constante de Boltzmann.
