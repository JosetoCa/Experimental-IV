## INTRODUCCIÓN

Debido a la radiación cósmica y a la presencia de materiales radioactivos en el planeta Tierra, los detectores de radiación perciben en cierto grado (dependiendo de la naturaleza y tamaño del detector) una cantidad de ruido en sus medidas, una radiación natural, no intencionada, denominada radiación de fondo [1]. La sensibilidad de un detector a la radiación de fondo determina el mínimo nivel de radiación que es capaz de detectar. Por ello se vuelve importante tener una medida de esta radiación, entenderla y saberla contrarrestar. La principal fuente de la radiación de fondo es la radiactividad terrestre [2], es decir, partículas radioactivas que han estado en el núcleo terrestre desde su formación y que decaen hasta alcanzar la superficie. Con base en [3], el elemento preponderante en la superficie es el radón-222, un gas que se desintegra en partículas alfa. Este gas, como otros nucleidos radiactivos, se desintegra por decaimiento radioactivo, un cambio espontáneo de un núcleo atómico inestable que pierde energía y que produce un nuevo núcleo o simplemente un cambio en su energía. El decaimiento es un proceso aleatorio donde no se puede predecir el instante exacto en que un átomo decaerá, pero se tiene una constante de decaimiento, la probabilidad por unidad de tiempo de que un núcleo decaiga [4]. Gracias a ello, se sabe que el proceso sigue una distribución exponencial. Además, si en lugar de considerar el instante de decaimiento como la variable aleatoria, y nos concentramos en el número de átomos que decaerán en un lapso, estaremos ante un proceso de conteo que sigue una distribución de Poisson [5].

Una forma de medir dicha radiación es por medio de un tubo escintilador de NaI(Tl) --Yoduro de sodio con impureza de Talio-- que consiste en un cristal que produce destellos cuando la radiación ionizante lo atraviesa. Los primeros materiales sólidos utilizados para la detección de partículas eran tubos escintiladores; por ejemplo, en 1910, en el experimento de Rutherford, se contaban los destellos en una pantalla de sulfuro de zinc que se producían al impactar partículas alfa en ella, los cuales se observaban por medio de un microscopio [3]. Dado que estos destellos son muy tenues, el tubo está acoplado a un fotomultiplicador, que los amplifica y produce un pulso eléctrico intenso a la salida.


Los elementos radiactivos experimentan constantemente desintegraciones nucleares medibles, cuya naturaleza depende del isótopo en cuestión. Por tanto, es posible medir espectros de radiación característicos de distintos elementos, los cuales se diferencian de la radiación de fondo. Un ejemplo es el americio-241, un isotopo radiactivo sintético usado en algunos detectores de humo y que puede ser encontrado en la naturaleza en forma de polvo microscópico en la tierra, plantas y agua, proveniente de pruebas atómicas [6], cuya actividad radioactiva sigue un proceso probabilístico compatible con la distribución de Poisson, empleando pruebas de bondad $\chi^2$ [7]. Otro ejemplo es el tungsteno toriado usado en soldadura de arco TIG, y que debido a su contenido de torio, emite cierta cantidad de radiación [8], cuyos conteos de desintegración también siguen una estadística de Poisson [9].

En el presente trabajo, empleando el tubo escintilador mencionado, se verificará que la radiación de fondo sigue una distribución de Poisson por medio de una prueba de bondad $\chi^2$. Además, usando la misma prueba de bondad, se comparará el espectro radiactivo medido por el tubo escintilador con la presencia de una muestra de americio-241 y una barra de soldadura por arco con un contenido de 2% de torio según la literatura. ## DESARROLLO Y METODOLOGÍA




### MARCO TEÓRICO

- Decaimiento radiactivo

Primero, es importante entender cómo funciona el decaimiento radiactivo matemáticamente hablando. Tomando como referencia [3], físicamente, este decaimiento representa un cambio espontáneo del núcleo o de sus energías, del que no se tiene certeza sobre cuándo y a qué razón se hará, pero se puede razonar con una probabilidad de que el núcleo decaiga en un cierto periodo de tiempo. Esta probabilidad se caracteriza por medio de la constante de decaimiento $\lambda$, que, para algunos isótopos, cumple que es independiente del número de núcleos y su edad, y es la misma para todos los núcleos de la muestra.
Matemáticamente esta constante permite establecer una ley o fórmula que establece que la variación del número de núcleos en una muestra de masa $m$ de un isótopo dado es proporcional a dicho número multiplicado por la constante de decaimiento:

$$\frac{d N(t)}{dt} = - \lambda N(t)$$

La solución a esta ecuación es una curva exponencial de la forma $N(t) = N_0 e^{- \lambda t}$, donde la parte exponencial nos dice la probabilidad de que un atomo no decaiga hasta un tiempo $t$.

- Prueba de bondad $\chi^2$

Con base en la referencia [10]. Si se ha realizado una cantidad de medidas de la variable aleatoria $x$, sobre la cual, a priori (por razones teóricas o experimentales), se puede esperar siga cierta distribución, existe una forma de saber si hay razones estadísticas de peso para darle validez a esta suposición. De la medida, los $N$ datos tomados se deben dividir en $n$ intervalos de los posibles valores que toma $x$ y contar por cada intervalo, el numero de medidas $O_k$ que caen en el $k$-esimo intervalo construido. Debido a que se asumió inicialmente cierta distribución para la variable, se debe obtener un número de medidas esperadas $E_k$ en el mismo $k$-esimo intervalo según dicha distribución. La herramienta que nos dará una prueba de bondad (o que tan cercana es la medida con la distribución esperada) sobre la variable $x$ y la distribución propuesta es la variable $\chi^2$ dada por

$$ \chi^2 = \sum^{n}_{k = 1}  \frac{ (O_k – E_k)^2}{E_k}$$

La cual sigue su propia distribución llamada de la misma forma y dada por

$$f_k (x) = \frac{1}{2^{\frac{k}{2}} \Gamma (\frac{k}{2})} x^{\frac{k}{2}-1}e^{-\frac{x}{2}}$$

Gracias al valor de $\chi^2$ y su distribución, se puede realizar una prueba de hipótesis dado un nivel de significancia (generalmente del 5%) sobre el valor obtenido para chi y el correspondiente al nivel de significancia para la distribución chi-cuadrado. Como la distribución de $\chi^2$ posee una variable o parámetro $k$, se deben reducir los grados de libertad de este parámetro por medio del número de parámetros auxiliares calculados a partir de los datos experimentales que sirvieron para caracterizar o determinar el modelo probabilístico esperado y, por ende, los valores esperados $E_k$. Si este numero es $c$, los grados de libertad (y el parámetro en el que se debe evaluar la distribución de $\chi^2$) será $k-c-1$ (detalles adicionales en la ref. [10].). Así, al comparar los valores del chi-cuadrado de los datos y el asociado al nivel de significancia, se puede determinar el peso de la suposición sobre la naturaleza aleatoria que siguen los datos experimentales.

Por supuesto, los valores observados tienen una incertidumbre, si siguieran una distribución de Poisson la incertidumbre del valor observado N es $\sqrt{N}$, eso sin analizar fuentes de incertidumbre adicionales como el proceso de detección de un conteo con el tubo escintilador. Luego, para la prueba de bondad se desarrolló el siguiente ajuste por medio de la función curve_fit de la librería scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import poisson

#Primero, utilizamos la función unique de numpy que me entrega dos "arrays", uno con los datos que se repiten, otro con cuántas veces se repiten,
#Un problema con este método es que cuando aparece un valor, como en nuestro caso el 7, no aparece tampoco en los arrays


valores, frecuencias = np.unique(conteos_por_intervalo, return_counts=True)

#Para corregir ese detalle, se crean dos nuevos "arrays" y se usa

valores_ = np.arange(min(valores), max(valores)+1)
frecuencias_ = np.zeros(len(valores_))

#np.searchsorted me devuelve un arreglo con los índices donde insertar
# cada valor del arreglo de np.unique para que valores_ permanezca ordenado (de menor a mayor)

indices = np.searchsorted(valores_, valores)
frecuencias_[indices] = frecuencias



# Definimos la función de Poisson a ajustar, con una amplitud correspondiente
def poisson_pmf(k, lamb, amp):
    return amp*poisson.pmf(k, lamb)

# Ajuste con curve_fit
params_opt, params_cov = curve_fit(poisson_pmf, valores_, frecuencias_)

# Resultado del ajuste
lambda_ajustada = params_opt[0]
amp_ajustada = params_opt[1]
print('Parámetros ajustados:')
print(f"λ ajustada = {lambda_ajustada:.3f} +- {np.sqrt(params_cov[0,0]):.3f}")
print(f"amplitud ajustada = {amp_ajustada:.3f} +- {np.sqrt(params_cov[1,1]):.3f}")



# Graficamos
x = np.arange(min(valores_), max(valores_)+1)
plt.bar(valores_, frecuencias_, alpha=0.6, label='Datos')
plt.plot(x, poisson_pmf(x, lambda_ajustada, amp_ajustada), 'o-', color='red', label=f'Poisson ajustada (λ={lambda_ajustada:.2f})')

plt.xlabel("Número de eventos")
plt.ylabel("Probabilidad")
plt.title("Ajuste de distribución de Poisson")
plt.legend()
plt.grid()
plt.show()

print(f'Media del ajuste: {lambda_ajustada:.2f}')
print(f'Varianza del ajuste: {lambda_ajustada:.2f}')
print(f'Desviación estándar del ajuste: {np.sqrt(lambda_ajustada):.1f}')
```

La incertidumbre de cada parámetro del ajuste la consideramos igual a la raíz cuadrada de su covarianza. Así el k-ésimo valor esperado tendrán una forma:

$$
E_k = \frac{a\lambda^k e^{-\lambda}}{k!}
$$


con $\lambda = (4.1 \pm 0.4)$ y $a = (30 \pm 4) $. Después se hizo la puebra de bondad para los datos observados y el ajuste, teniendo en cuenta que los datos, siendo discretos, están organizados en 9 intervalos, y con dos ligaduras, son 6 los grados de libertad para la prueba de Bondad, testeando la hipótesis: nuestros datos siguen la distribución dada por el ajuste de Poisson.


```python
from scipy.stats import chi2

# Definimos una función para encontrar el valor de chi^2 de los conteos y el ajuste.
def chi_sqr(dathistexp,dathistaj):
    return sum(((dathistexp-dathistaj)**2)/(dathistaj))


chi=chi_sqr(frecuencias_, poisson_pmf(valores_, lambda_ajustada, amp_ajustada))
print(chi)

# Definimos los grados de libertad: -3: datos finitos (30), dos ligaduras en el ajuste: la amplitud y la media para la distribución de Poisson.
df = len(frecuencias_)-3

# El nivel de significancia para nuestra prueba en un valor estándar 0.05

alpha = 0.05

# Graficamos histogramas con nuestros datos y los valores del ajuste

plt.bar(valores_, frecuencias_, width=np.ones(len(valores_)), edgecolor='black',label='Observado')
plt.bar(valores_, poisson_pmf(valores_, lambda_ajustada, amp_ajustada), width=np.ones(len(valores_)), edgecolor='red',alpha=0.5,label='Esperado')

plt.xlabel("Conteos")
plt.ylabel("Frecuencia")
plt.title("Histogramas para la prueba de bondad")
plt.grid(True)
plt.legend()
plt.show()



#Encontramos el valor de la distribución de chi^2 para el cual rechazamos o aceptamos nuestra prueba de hipótesis (normalmente se llama p = 1-alfa)
chi2_critico = chi2.ppf(1-alpha, df)
print(f"Valor crítico de chi2 para {df} grados de libertad y alpha={alpha}: {chi2_critico:.2f}")
if chi < chi2_critico:
    print("No se rechaza la hipótesis nula: estadísticamente, no hay que me diga que los datos no siguen una distribución de Poisson.")
else:
    print("Se rechaza la hipótesis nula: los datos, estadísticamente, no siguen una distribución de Poisson.")
```

El valor de $ \chi^2$ fue de $5.03$

La prueba de bondad está pensada en testear la hipótesis de que los datos siguen la misma distribución del ajuste dado por la radiación de fondo.

```python
from scipy.stats import chi2


conteos_americio = np.array([1, 3, 5, 5, 7, 4, 8, 5, 3, 8, 8, 5, 4, 0, 1, 6, 8, 4, 8, 4, 3, 6, 4, 2, 5, 7, 2, 4, 1, 5])

def chi_sqr(dathistexp,dathistaj):
    return sum(((dathistexp-dathistaj)**2)/(dathistaj))

valores_americio, frecuencias_americio = np.unique(conteos_americio, return_counts=True)



valores_americio_ = np.arange(min(valores_americio), max(valores_americio)+1)
frecuencias_americio_ = np.zeros(len(valores_americio_))



indices = np.searchsorted(valores_americio_, valores_americio)
frecuencias_americio_[indices] = frecuencias_americio



chi=chi_sqr(frecuencias_americio_, poisson_pmf(valores_americio_, lambda_ajustada, amp_ajustada))
print(chi)



df = len(frecuencias_americio_)-1
alpha = 0.05
plt.bar(valores_americio_, frecuencias_americio_, width=np.ones(len(valores_americio_)), edgecolor='black',label='Observado')
plt.bar(valores_americio_, poisson_pmf(valores_americio_, lambda_ajustada, amp_ajustada), edgecolor='red',alpha=0.5,label='Esperado')
plt.xlabel("Nivel de gris")
plt.ylabel("Frecuencia")
plt.title("Histogramas para la prueba de bondad")
plt.grid(True)
plt.legend()
plt.show()



chi2_critico = chi2.ppf(1-alpha, df)
print(f"Valor crítico de chi2 para {df} grados de libertad y alpha={alpha}: {chi2_critico:.2f}")
if chi < chi2_critico:
    print("No se rechaza la hipótesis nula: estadísticamente, no hay que me diga que los datos no siguen una distribución de Poisson como la radiación de fondo.")
else:
    print("Se rechaza la hipótesis nula: los datos, estadísticamente, no siguen una distribución de Poisson como la radiación de fondo.")
```

### CONCLUSIONES
Desde una perspectiva cualitativa, comparando la media y la varianza, los conteos de la radiación de fondo ya sugieren un comportamiento compatible con una distribución de Poisson. Luego, los parámetros óptimos del ajuste para un modelo de Poisson hecho sobre las medidas experimentales mediante la función curve_fit, $\lambda = (4.1 \pm 0.4)$ y $a = (30 \pm 4) $, se aproximan notablemente a los valores teóricos esperados para una distribución de este tipo con $a = 30$ medidas y una media $\lambda = 4$. Por último, la prueba de hipótesis, con un valor de $\chi ^2 = 5.03$, claramente dentro del rango de aceptación como se puede apreciar en el análisis, da una base estadística sólida para afirmar que la detección de la radiación de fondo por medio de un tubo escintilador puede modelarse adecuadamente con una distribución de Poisson con los parámetros obtenidos.

Al analizar los datos obtenidos para las muestras de americio y torio, las pruebas de hipótesis, con valores de $\chi^2_{\text{Am}} = 20.46$ y $\chi^2_{\text{Th}} = 39.33$ respectivamente, las cuales están considerablemente fuera del rango de aceptación (ver Análisis de datos), son contundentes en afirmar que, estadísticamente, los conteos registrados en presencia de las muestras no siguen la misma distribución de Poisson que la radiación de fondo.

Si bien la literatura señala que los procesos de desintegración del americio-241 y del torio en soldadura TIG responden individualmente a modelos probabilísticos tipo Poisson, los resultados del presente trabajo muestran que sus distribuciones son distintas a la que sigue la radiación de fondo. Es decir, el tubo escintilador detecta una señal que puede atribuirse a la presencia de las muestras.

En conjunto, los resultados respaldan el comportamiento esperado según la literatura. Sin embargo, una posible mejora en la precisión del experimento consistiría en automatizar y mejorar la toma de datos durante el procedimiento experimental. En particular, si se lograra una mayor exactitud y consistencia en la medición del tiempo en el que se alcanzan cuatro conteos, y pudiendo detener la detección en el momento justo en que el cronómetro marca el tiempo medio para llegar a cuatro conteos, se podría reducir la incertidumbre asociada tanto al tiempo como a los conteos, mejorando así la confiabilidad de las mediciones de la radiación de fondo y las muestras posiblemente radiactivas.
